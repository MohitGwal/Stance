```python
def remove_punctuations(text):
    punctuations = string.punctuation
    return text.translate(str.maketrans('', '', punctuations))
df1['clean_text1'] = df1['clean_text1'].apply(lambda x: remove_punctuations(x))
df1.head()
```

```python
import re
def remove_spl_chars(text):
    text = re.sub('[^a-zA-Z0-9]', ' ', text)
    text = re.sub('\s+', ' ', text)
    return text
df1['clean_text1'] = df1['clean_text1'].apply(lambda x: remove_spl_chars(x))
df1.head()
```

```python
X1 = df1['clean_text1'].values

embeddings1=[]
```

```python
from simpletransformers.language_representation import RepresentationModel
        
Tweet1=(X1.tolist())
tweet1=Tweet1[:40]

model = RepresentationModel(
        model_type="bert",
        model_name="bert-base-uncased",
        use_cuda=False
    )
for i in tweet1:  
  word_vectors = model.encode_sentences(i, combine_strategy=None)
  word_vectors=np.array(word_vectors)
  embeddings1.append(word_vectors)
```

```python
import numpy as np

embeddings_3d=embeddings1

max_sentence_length = max(embedding_3d.shape[0] for embedding_3d in embeddings_3d)
max_word_length = max(embedding_3d.shape[1] for embedding_3d in embeddings_3d)
embedding_dimension = embeddings_3d[0].shape[2]

for i in range(len(embeddings_3d)):
    embedding_3d = embeddings_3d[i]
    if embedding_3d.shape[0] < max_sentence_length:
        zero_padding = np.zeros((max_sentence_length - embedding_3d.shape[0], max_word_length, embedding_dimension))
        embeddings_3d[i] = np.concatenate((embedding_3d, zero_padding), axis=0)
    if embedding_3d.shape[1] < max_word_length:
        zero_padding = np.zeros((max_sentence_length, max_word_length - embedding_3d.shape[1], embedding_dimension))
        embeddings_3d[i] = np.concatenate((embedding_3d, zero_padding), axis=1)

concatenated_embeddings1 = np.concatenate(embeddings_3d)

num_sentences = len(embeddings_3d)
concatenated_embeddings1 = concatenated_embeddings1.reshape((num_sentences, max_sentence_length, max_word_length, embedding_dimension))

print(concatenated_embeddings1)
```
